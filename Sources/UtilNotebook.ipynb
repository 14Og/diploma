```python
names = linear.stats_dataframe["Panel"].to_list()
x = np.arange(len(names)) * 2
width = 0.4
multiplier = 2
errors = {"Linear":linear.stats_dataframe["mean_error_linearized"], 
        "Geometrical (linear offsets)":geom.stats_dataframe["mean_error_linear_offsets"], 
        "Polynomial (cubic)":poly.stats_dataframe["mean_error_cube"],
        "GBDT": gbdt.stats_dataframe["mean_error_boosted"],
        "MLP": mlp.stats_dataframe["mean_error_mlp"]}

colors=["green", 'blue', 'orange', "grey", "red"]
import matplotlib.pyplot as plt
fig, ax = plt.subplots(layout='constrained', figsize=(30,12))
for (calib_type, error_levels), color in zip(errors.items(), colors):
    offset = width * multiplier
    rects = ax.bar(x + offset, error_levels, width, label=calib_type, facecolor='none', edgecolor=color, linewidth=2)
    # ax.bar_label(rects, padding=5, fmt="%.2f")
    multiplier += 1


ax.set_xlabel("Sensor id")
ax.set_ylabel("Mean error level")
ax.set_title("Calibration algorithms comparison")
ax.set_xticks(x + width, names)
ax.legend(loc='upper left', ncols=3)
plt.show()
```
